````python
# Montando uma tabela dos tipos de renda por data
tipo_renda_counts = df.pivot_table(index='data_ref', columns='tipo_renda', aggfunc='size', fill_value=0)
tipo_renda_counts
````

````python
ax = tipo_renda_counts.plot(kind='bar', rot=90)
ax.set_xlabel('Data de Referência')
ax.set_ylabel('Contagem')
ax.set_title('Contagem de Tipos de Renda ao Longo do Tempo')
ax.legend(bbox_to_anchor=(1, 1))
````
